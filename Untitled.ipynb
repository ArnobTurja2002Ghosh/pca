```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the cancer dataset.

```python
df = pd.read_csv("cancer_data.csv")
#df.rename(columns={'Height(Inches)': 'height', 'Weight(Pounds)': 'weight'}, inplace=True)
#df.weight = df.weight * 1.25 # artificially making everyone weigh 1.25 times because no one was overweight originally lol
df
```

```python
df.isnull().sum()
```

Ensure that you use only numerical data for PCA

```python
# One-hot encode the 'Sex' column
df = pd.get_dummies(df, columns=['diagnosis'], drop_first=True)

# Display the first few rows of the encoded dataset
df
```

Scale the data so that all features have a similar range

```python
from sklearn.preprocessing import StandardScaler

# Select the features to standardize
features_to_scale = df.columns

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform the data
abalone_data_scaled = scaler.fit_transform(df[features_to_scale])

# Convert the scaled data back to a DataFrame for better readability
abalone_data_scaled_df = pd.DataFrame(abalone_data_scaled, columns=features_to_scale)

# Display the first few rows of scaled data
abalone_data_scaled_df
```

2. PCA Application:
- Perform PCA on the prepared data.

3. Variance Analysis:
- Calculate and display the amount of variance captured by each of the selected principal
components.
- Present the cumulave variance explained by the principal components. 

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

# Initialize PCA
pca = PCA(n_components=10)

# Fit the scaled data
pca.fit(abalone_data_scaled)

# Get the explained variance ratio
explained_variance = pca.explained_variance_ratio_
cumulative_variances = np.cumsum(explained_variance)

# Plot the explained variance
plt.figure(figsize=(10, 6))
plt.bar(range(1,1+len(explained_variance)), explained_variance, alpha=0.5, align='center', label='Individual Explained Variance')
line = plt.step(range(1,1+len(explained_variance)), cumulative_variances, linestyle='-', label='Cumulative Explained Variance', where='mid')
plt.ylabel('Explained Variance Ratio')
plt.xlabel('Principal Component')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import PCA

# Initialize PCA and specify the number of principal components to keep
pca = PCA(n_components=10)

# Apply PCA to the scaled features
X_pca = pca.fit_transform(abalone_data_scaled)

np.max(X_pca[:,1])
```

4. Visualizaon:
- Create a 3D scaer plot using the first three principal components as axes.
- Use the diagnosis as the hue to differenate between the classes.

```python
from matplotlib.colors import ListedColormap
fig = plt.figure(figsize=(9, 6), layout='constrained')

ax = fig.add_subplot(projection="3d")

scatter=ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=abalone_data_scaled_df['diagnosis_M'], cmap = ListedColormap(['b', 'r']), marker='X')

ax.set_title('3D PCA PLOT')

ax.set_xlabel("Principal Component 1")

ax.set_ylabel("Principal Component 2")
ax.set_yticks([-7.5, -5, -2.5, 0, 2.5, 5, 7.5, 10, 12.5])

ax.set_zlabel("Principal Component 3")
ax.set_zticks([-4, -2, 0, 2, 4, 6, 8, 10])

L=ax.legend(scatter.legend_elements()[0], ['Malignant', 'Benign'], loc="upper right")

```

 create a 2D scatter plot using the first two principal components as axes with
diagnosis as the hue

```python
from matplotlib.colors import ListedColormap
fig = plt.figure()

ax = fig.add_subplot()

scatter=ax.scatter(X_pca[:, 0], X_pca[:, 1], c=abalone_data_scaled_df['diagnosis_M'], cmap = ListedColormap(['b', 'r']), marker='X', linewidth=0)

ax.set_title('2D PCA PLOT')

ax.set_xlabel("Principal Component 1")

ax.set_ylabel("Principal Component 2")
ax.set_yticks([-7.5, -5, -2.5, 0, 2.5, 5, 7.5, 10, 12.5])

L=ax.legend(scatter.legend_elements()[0], ['M', 'B'], loc="upper right")
```
